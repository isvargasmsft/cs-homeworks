*Indefinido*

## Pregunta 2

En general, si $u \in R^n$ y $g \in R^m$, y $p=1$ es el paso de la convolucion, que dimensiones tiene h, suponiendo que no hay un padding de 0's en ninguno de los dos operandos de la convolucion.

Tenemos que:

$ h[n] = f[n] \cdot g[n] = \sum_{-\infty}^{\infty} f[m] \cdot g[n-m] $

Para que la convolución sea válida, $u$ y $g$ deben tener índices de tamaño $n$, de modo que el primer valor de $n$ es $0$ y el último es $n-m$.

La longitud de $h$ es el número de valores que puede tomar $m$ entre $0$ y $n - m$. Por lo tanto:

$ \text{longitud de } h = n - m + 1 $

Por lo tanto, la dimensión de $ h $ es igual a $\mathbb{R}^{(n - m + 1)} $.

## Pregunta 3

Demuestre la siguiente propiedad: $g(x) * (\delta(x) + k \delta(x-s)) = g(x) + k g(x-s)$

La función $\delta$ de Dirac tiene las siguientes propiedades:

1. $\delta(x) = 0$ para $x \neq 0$

1. $\int_{-\infty}^{\infty} \delta(x) , dx = 1$.

1. $\delta(x-a) \cdot f(x) = f(a) \cdot \delta(x-a)$ para cualquier función $f(x)$.

Ahora, tenemos que:

$g(x) \cdot (\delta(x) + k \cdot \delta(x-s))$

Redistribuimos usando la tercera propiedad:

$ g(x) \cdot \delta(x) + g(x) \cdot k \cdot \delta(x-s) $

$ g(0) \cdot \delta(x) + k \cdot g(s) \cdot \delta(x-s)$

Sustituimos variables:

$ g(x) \cdot \delta(x) + k \cdot g(x-s) \cdot \delta(x-s) $

$g(x) + k \cdot g(x-s)$

Por lo tanto:

$g(x) \cdot (\delta(x) + k \cdot \delta(x-s)) = g(x) + k \cdot g(x-s) $
